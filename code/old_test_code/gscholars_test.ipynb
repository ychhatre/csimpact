```python
scholar_urls = []
for i in range(data.shape[0]):
    if(data.iat[i,3] != 'NOSCHOLARPAGE'):
        scholar_urls.append(f'https://scholar.google.com/citations?hl=en&user={data.iat[i,3]}')
    else:
        scholar_urls.append('')
```

```python
citations_list = []

h_index_list = []
```

```python
for i in range(data.shape[0]):
    
    if data.iloc[i,3] == 'NOSCHOLARPAGE':
        citations_list.append('NOSCHOLARPAGE')
        h_index_list.append('NOSCHOLARPAGE')
    else:
        gscholars_text = requests.get(scholar_urls[i]).text
        soup = BeautifulSoup(gscholars_text, 'lxml')
        values = soup.find_all('td', class_ = 'gsc_rsb_std')
        if values == []:
            citations_list.append('NODATAFOUND')
            h_index_list.append('NODATAFOUND')
            print(f'{i} none')
            continue
        citations = values[0].text
        h_index = values[2].text
        citations_list.append(citations)
        h_index_list.append(h_index)
    print(i)
```

```python
#remove all abbreviations


name1 = 'Kevin S.     '

def edit_name(name):
    edited_name = ''
    name = name.split()
    for word in name:
        if '.' in word:
            name.remove(word)
            
        if len(name) == 1:
            edited_name = name[0]
        else:
             edited_name = name[0] + ' '+ name[-1]
    
    return edited_name

new_name = edit_name(name1)
print(new_name)
```
