```python
scraped_data = pd.read_csv(os.getcwd().replace('code','') + f'/data/replaced_bad_prof_data_12000_20650.csv')
```

```python
scraped_data = scraped_data.drop(['Unnamed: 0'], axis=1)
```

```python
new_data = old_data.copy()
```

```python
new_data
```

```python
for i in range(scraped_data.shape[0]):
    row = scraped_data.loc[i]
    row_to_replace = new_data.loc[(new_data['name']==row['Name']) & (new_data['affiliation'] == row['affiliation'])]
    index = row_to_replace.index.tolist()[0]
    new_data.loc[index] = row.tolist()
```
