```python
#clean up the names by removing zeros from the end (for example A. P. Vinod 0001)
for i in range(data_8.shape[0]):
    temp = ''
    if '0' in data_8.iloc[i,0]:
        temp = data_8.iloc[i,0]
        temp = temp.split()
        temp.remove(temp[-1])
        data_8.iloc[i,0] = ' '.join(str(x) for x in temp)
```
